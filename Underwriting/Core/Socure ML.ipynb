```python
def adj_r2_score(r2, n, k):
    return 1-((1-r2)*((n-1)/(n-k-1)))
```

## Clustering

```python
def plot_embedding(X, y, title=None, num=10):
    x_min = np.min(X, 0)
    x_max = np.max(X, 0)
    X = (X-x_min)/(x_max-x_min)
    
    plt.figure(figsize=(18, 9))
    plt.scatter(X[:,0], X[:,1], color=plt.cm.Set1(y/num))
    
    plt.xticks([])
    plt.yticks([])
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
from sklearn.manifold import TSNE
subSpace = TSNE().fit_transform(final.values)
```

```python
from sklearn.cluster import KMeans
cluster = KMeans(n_clusters=4, max_iter= 1000, n_init=100).fit(final.values)
plot_embedding(subSpace, cluster.labels_, 'K-Means projection',4)
```

```python
clus = pd.DataFrame(cluster.labels_, columns=['label'])
sns.set(rc={'figure.figsize':(15,6)})
ax = sns.countplot(x="label", data=clus)
ax.set_title("Cluster Distribution", fontsize=24)
ax.set_xlabel("Cluster #",fontsize=20)
ax.set_ylabel("Number of Users",fontsize=20)
ax.tick_params(labelsize=16)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/(2),height + .2,'{:1.0f}'.format(height),ha="center",fontsize=18)
plt.show()
```

```python
def plotScatterMatrix(df, plotSize, textSize):
    df = df.select_dtypes(include =[np.number]) # keep only numerical columns
    # Remove rows and columns that would lead to df being singular
    df = df.dropna('columns')
    df = df[[col for col in df if df[col].nunique() > 1]] # keep columns where there are more than 1 unique values
    columnNames = list(df)
    if len(columnNames) > 30: # reduce the number of columns for matrix inversion of kernel density plots
        columnNames = columnNames[:30]
    df = df[columnNames]
    ax = pd.plotting.scatter_matrix(df, alpha=0.75, figsize=[plotSize, plotSize], diagonal='kde')
    corrs = df.corr().values
    for i, j in zip(*plt.np.triu_indices_from(ax, k = 1)):
        ax[j, i].annotate('%.3f' % corrs[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=textSize)
    #plt.suptitle('Scatter and Density Plot', fontsize=20)
    n = len(df.columns)
    for x in range(n):
        for y in range(n):
            # to get the axis of subplots
            axs = ax[x, y]
            # to make x axis name vertical  
            axs.xaxis.label.set_rotation(90)
            # to make y axis name horizontal 
            axs.yaxis.label.set_rotation(0)
            # to make sure y axis names are outside the plot area
            axs.yaxis.labelpad = 50
    plt.show()
```

## ML Predictions

### Joust Score 1

```python
tresh = 5
atts = []
for k in range(len(ascorerank)):
    if (ascorerank[k] <= tresh):
        atts.append(cross_x.columns.values[k])
atts
```

```python
reg = LinearRegression()
reg = XGBRegressor(objective='reg:squarederror')
cross_xsc = cross_x[atts]
score = cross_val_score(reg, cross_xsc.values, cross_y.joust_status.values, cv=5, scoring='r2')
print(sum(score)/5)
```

```python
train_y2 = cross_y.copy(deep=True)
train_y2['joust_score'] = cross_val_predict(reg, cross_xsc.values, cross_y.joust_status.values, cv=10)
train_y2.loc[train_y2['joust_score'] > 2, 'joust_score'] = 2
train_y2.loc[train_y2['joust_score'] < 0, 'joust_score'] = 0
train_y2['joust_status'].plot(figsize=(15,7), fontsize=14, label='Joust Status')
train_y2['joust_score'].plot(figsize=(15,7), fontsize=14, label='Joust Score')
plt.figtext(.5,.9,'Linear Regression',fontsize=22,ha='center')
plt.legend(loc='lower left')
plt.show()
print('adj overall r^2 = ' + str(round(adj_r2_score((r2_score(train_y2.joust_status, train_y2.joust_score)),train_y2.shape[0], train_y2.shape[1]),4)))
print('overall mse = ' + str(round(mean_squared_error(train_y2.joust_status, train_y2.joust_score),4)))
```
