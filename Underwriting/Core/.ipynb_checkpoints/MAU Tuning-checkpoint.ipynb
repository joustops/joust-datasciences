```python
def adj_r2_score(r2, n, k):
    return 1-((1-r2)*((n-1)/(n-k-1)))
```

## Clustering

```python
def plot_embedding(X, y, title=None, num=10):
    x_min = np.min(X, 0)
    x_max = np.max(X, 0)
    X = (X-x_min)/(x_max-x_min)
    
    plt.figure(figsize=(18, 9))
    plt.scatter(X[:,0], X[:,1], color=plt.cm.Set1(y/num))
    
    plt.xticks([])
    plt.yticks([])
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
from sklearn.manifold import TSNE
subSpace = TSNE().fit_transform(final.values)
```

```python
from sklearn.cluster import KMeans
cluster = KMeans(n_clusters=10, max_iter= 1000, n_init=100).fit(final.values)
plot_embedding(subSpace, cluster.labels_, 'K-Means projection',10)
```

```python
outl = pd.DataFrame(final.index.values, columns=['id'])
outl['x-coor'] = subSpace[:,0]
outl['y-coor'] = subSpace[:,1]
print("Naïve Outliers")
print(outl[outl['x-coor'] == outl['x-coor'].max()])
print(outl[outl['x-coor'] == outl['x-coor'].min()])
print(outl[outl['y-coor'] == outl['y-coor'].max()])
print(outl[outl['y-coor'] == outl['y-coor'].min()])
```

```python
clus = pd.DataFrame(cluster.labels_, columns=['label'])
sns.set(rc={'figure.figsize':(15,6)})
ax = sns.countplot(x="label", data=clus)
ax.set_title("Cluster Distribution", fontsize=24)
ax.set_xlabel("Cluster #",fontsize=20)
ax.set_ylabel("Number of Users",fontsize=20)
ax.tick_params(labelsize=16)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/(2),height + .2,'{:1.0f}'.format(height),ha="center",fontsize=18)
plt.show()
```

## Analysis

```python
def plotScatterMatrix(df, plotSize, textSize):
    df = df.select_dtypes(include =[np.number]) # keep only numerical columns
    # Remove rows and columns that would lead to df being singular
    df = df.dropna('columns')
    df = df[[col for col in df if df[col].nunique() > 1]] # keep columns where there are more than 1 unique values
    columnNames = list(df)
    if len(columnNames) > 50: # reduce the number of columns for matrix inversion of kernel density plots
        columnNames = columnNames[:50]
    df = df[columnNames]
    ax = pd.plotting.scatter_matrix(df, alpha=0.75, figsize=[plotSize, plotSize], diagonal='kde')
    corrs = df.corr().values
    for i, j in zip(*plt.np.triu_indices_from(ax, k = 1)):
        ax[j, i].annotate('%.3f' % corrs[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=textSize)
    #plt.suptitle('Scatter and Density Plot', fontsize=20)
    n = len(df.columns)
    for x in range(n):
        for y in range(n):
            # to get the axis of subplots
            axs = ax[x, y]
            # to make x axis name vertical  
            axs.xaxis.label.set_rotation(90)
            # to make y axis name horizontal 
            axs.yaxis.label.set_rotation(0)
            # to make sure y axis names are outside the plot area
            axs.yaxis.labelpad = 50
    plt.show()
```

## Joust Status

```python
model = XGBRegressor(objective='reg:squarederror')
#model = LinearRegression()
rfe = RFE(model, 1)
reg = rfe.fit(cross_x.values, cross_y.joust_status.values)
print(reg.ranking_)
```

```python
ranks = pd.DataFrame((cross_x.columns.values), columns=['Attribute'])
ascorerank = reg.ranking_
ranks['Ranking'] = ascorerank
ranks = ranks.sort_values('Ranking', ascending=True)
ranks = ranks.set_index('Ranking')
ranks
```

## ML Predictions

### via Regression

```python
tresh = 5
atts = []
for k in range(len(ascorerank)):
    if (ascorerank[k] <= tresh):
        atts.append(cross_x.columns.values[k])
print(atts)
```

```python
#reg = LinearRegression()
reg = XGBRegressor(objective='reg:squarederror')
cross_xsc = cross_x[atts]
score = cross_val_score(reg, cross_xsc.values, cross_y.joust_status.values, cv=5, scoring='r2')
print(sum(score)/5)
```

```python
train_y2 = cross_y.copy(deep=True)
train_y2['joust_score'] = cross_val_predict(reg, cross_xsc.values, cross_y.joust_status.values, cv=5)
train_y2.loc[train_y2['joust_score'] > 100, 'joust_score'] = 100
train_y2.loc[train_y2['joust_score'] < 0, 'joust_score'] = 0
train_y2['joust_status'].plot(figsize=(25,7), fontsize=14, label='Joust Status')
train_y2['joust_score'].plot(figsize=(25,7), fontsize=14, label='Joust Score')
plt.figtext(.5,.9,'Linear Regression',fontsize=22,ha='center')
plt.legend(loc='lower left')
plt.show()
print('adj overall r^2 = ' + str(round(adj_r2_score((r2_score(train_y2.joust_status, train_y2.joust_score)),train_y2.shape[0], train_y2.shape[1]),4)))
print('overall mse = ' + str(round(mean_squared_error(train_y2.joust_status, train_y2.joust_score),4)))
```

### via Classification

```python
tresh = 4
atts = []
for k in range(len(ascorerank)):
    if (ascorerank[k] <= tresh):
        atts.append(cross_x.columns.values[k])
print(atts)
```

```python
#reg = LogisticRegression(solver='liblinear')
reg = XGBClassifier()
cross_xst = cross_x[atts]
score = cross_val_score(reg, cross_xst.values, cross_y.joust_status.values, cv=5, scoring='accuracy')
print(sum(score)/5)
```

```python
train_y2 = cross_y.copy(deep=True)
train_y2['joust_score'] = cross_val_predict(reg, cross_xst.values, cross_y.joust_status.values, cv=5)
train_y2.loc[train_y2['joust_score'] > 100, 'joust_score'] = 100
train_y2.loc[train_y2['joust_score'] < 0, 'joust_score'] = 0
train_y2['joust_status'].plot(figsize=(35,7), fontsize=14, label='Joust Status')
train_y2['joust_score'].plot(figsize=(35,7), fontsize=14, label='Joust Score')
plt.figtext(.5,.9,'Logistic Regression',fontsize=22,ha='center')
plt.legend(loc='lower left')
plt.show()
print('overall accuracy = ' + str(accuracy_score(train_y2.joust_status, train_y2.joust_score)))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

mlaccuracy =  accuracy_score(train_y2.joust_status.astype('int').values, train_y2.joust_score.values) * 100
naiveacc = len(final[final['joust_status'] == 100])/len(final) * 100
print('\033[1m Accuracies: \033[0m')
print('The Naive accuracy is ' + str(round(naiveacc,2)) + '%.')
print('The ML accuracy is ' + str(round(mlaccuracy,2)) + '%.')

print('')

print('\033[1m Classification Report: \033[0m')
print(classification_report(train_y2.joust_status.astype('int').values, train_y2.joust_score.values))

print('\033[1m Confusion Matrix: \033[0m')
confusion_df = pd.DataFrame(confusion_matrix(train_y2.joust_status.astype('int').values, train_y2.joust_score.values),
                 columns=["Predicted Class " + str(class_name) for class_name in [1,2]],
                 index = ["Class " + str(class_name) for class_name in [1,2]])
acc_pct = [round(100*confusion_df['Predicted Class 1'].values[0]/(confusion_df['Predicted Class 2'].values[0]+confusion_df['Predicted Class 1'].values[0]),2), 
           round(100*confusion_df['Predicted Class 2'].values[1]/(confusion_df['Predicted Class 2'].values[1]+confusion_df['Predicted Class 1'].values[1]),2)]
confusion_df['Accuracy'] = acc_pct
print(confusion_df)
```

## Optimization

### via Regression

```python
lrscores = []
xgbscores = []
for tresh in range(len(ascorerank)):
    cvn = 5
    atts = []
    for k in range(len(ascorerank)):
        if (ascorerank[k] <= tresh+1):
            atts.append(cross_x.columns.values[k])
    reg = LinearRegression()
    cross_xsc = cross_x[atts]
    score = cross_val_score(reg, cross_xsc.values, cross_y.joust_status.values, cv=cvn, scoring='r2')
    lrscores.append(sum(score)/cvn)
    reg = XGBRegressor(objective='reg:squarederror')
    cross_xsc = cross_x[atts]
    score = cross_val_score(reg, cross_xsc.values, cross_y.joust_status.values, cv=cvn, scoring='r2')
    xgbscores.append(sum(score)/cvn)
res = pd.DataFrame(lrscores, columns=['lr score'])
res['xgb score'] = xgbscores
res
```

```python
print('best xgb score: ' + str(res[res['xgb score'] == max(res['xgb score'])]))
print('best linreg score: ' + str(res[res['lr score'] == max(res['lr score'])]))
```

```python
lrscores = []
xgbscores = []
for tresh in range(len(ascorerank)):
    atts = []
    cvn = 5
    for k in range(len(ascorerank)):
        if (ascorerank[k] <= tresh+1):
            atts.append(cross_x.columns.values[k])
    reg = LinearRegression()
    cross_xsc = cross_x[atts]
    train_y2 = cross_y.copy(deep=True)
    train_y2['joust_score'] = cross_val_predict(reg, cross_xsc.values, cross_y.joust_status.values, cv=cvn)
    train_y2.loc[train_y2['joust_score'] > 100, 'joust_score'] = 100
    train_y2.loc[train_y2['joust_score'] < 0, 'joust_score'] = 0
    lrscores.append(adj_r2_score((r2_score(train_y2.joust_status, train_y2.joust_score)),train_y2.shape[0], train_y2.shape[1]))
    reg = XGBRegressor(objective='reg:squarederror')
    train_y2 = cross_y.copy(deep=True)
    train_y2['joust_score'] = cross_val_predict(reg, cross_xsc.values, cross_y.joust_status.values, cv=cvn)
    train_y2.loc[train_y2['joust_score'] > 100, 'joust_score'] = 100
    train_y2.loc[train_y2['joust_score'] < 0, 'joust_score'] = 0
    xgbscores.append(adj_r2_score((r2_score(train_y2.joust_status, train_y2.joust_score)),train_y2.shape[0], train_y2.shape[1]))
res = pd.DataFrame(lrscores, columns=['lr score'])
res['xgb score'] = xgbscores
res
```

```python
print('best xgb score: \n' + str(res[res['xgb score'] == max(res['xgb score'])]))
print('best linreg score: \n' + str(res[res['lr score'] == max(res['lr score'])]))
```
